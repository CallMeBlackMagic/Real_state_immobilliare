```python
real_states_about = []
```

```python
def main_immobiare(req):
    source = req.content
    soup = BeautifulSoup(source, 'lxml')
    real_states = soup.find_all("li", {"class": "nd-list__item in-realEstateResults__item"})
    number_of_real_states = len(real_states)

    def get_real_state_info(real_state):
        price = real_state.contents[0].find('li', {"class": "nd-list__item in-feat__item in-feat__item--main in-realEstateListCard__features--main"}).text.strip()
        space = real_state.find('span', class_='in-mainFeatures__text')
        value_m_squared = space.find_previous_sibling(text=True).strip()
        real_states_about.append({"Price": price, "M**2": value_m_squared})
##########################################################################################        
        for i in range(number_of_real_states):
            get_real_state_info(real_states[i])        
```
